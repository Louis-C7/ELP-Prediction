```python
def get_ht_gsgc(playing_stat):
    team_list = playing_stat.groupby('HomeTeam').mean().T.columns
    htgs, htgc = dict(), dict()
    for i in team_list:
        ht = playing_stat[playing_stat['HomeTeam'] == i]
        htgs[i] = [sum(ht.iloc[0:idx, :]['FTHG']) for idx, _ in enumerate(ht.values)]
        htgc[i] = [sum(ht.iloc[0:idx, :]['FTAG']) for idx, _ in enumerate(ht.values)]
    HTGS, HTGC = list(), list()
    for idx, i in enumerate(playing_stat.HomeTeam):
        HTGS.append(htgs[i].pop(0))
        HTGC.append(htgc[i].pop(0))
    playing_stat['HTGS'] = HTGS
    playing_stat['HTGC'] = HTGC

def get_at_gsgc(playing_stat):
    team_list = playing_stat.groupby('AwayTeam').mean().T.columns
    atgs, atgc = dict(), dict()
    for i in team_list:
        at = playing_stat[playing_stat['AwayTeam'] == i]
        atgs[i] = [sum(at.iloc[0:idx, :]['FTAG']) for idx, _ in enumerate(at.values)]
        atgc[i] = [sum(at.iloc[0:idx, :]['FTHG']) for idx, _ in enumerate(at.values)]
    ATGS, ATGC = list(), list()
    for idx, i in enumerate(playing_stat.AwayTeam):
        ATGS.append(atgs[i].pop(0))
        ATGC.append(atgc[i].pop(0))
    playing_stat['ATGS'] = ATGS
    playing_stat['ATGC'] = ATGC
    
def get_gsgc(playing_stat):
    get_ht_gsgc(playing_stat)
    get_at_gsgc(playing_stat)
```

```python
for i in tqdm(range(0, 21)):
    get_gsgc(table[i])
```

```python
def get_points(playing_stat):
    team_list = playing_stat.groupby('HomeTeam').mean().T.columns
    ht_points, at_points = dict(), dict()
    for i in team_list:
        df = playing_stat[(playing_stat['HomeTeam'] == i) | (playing_stat['AwayTeam'] == i)]
        for index, row in df.iterrows():
            if (row['HomeTeam'] == i and row['FTR'] == 'H') or (row['AwayTeam'] == i and row['FTR'] == 'A'):
                df.loc[index, 'P'] = 3
            elif row['FTR'] == 'D':
                df.loc[index, 'P'] = 1
            else:
                df.loc[index, 'P'] = 0
        ans1, ans2 = list(), list()
        for idx, j in enumerate(df.values):
            if j[1] == i:
                ans1.append(sum(df.iloc[0:idx, :]['P']))
            elif j[2] == i:
                ans2.append(sum(df.iloc[0:idx, :]['P']))
            else:
                continue
        ht_points[i] = ans1
        at_points[i] = ans2
    HTP, ATP = list(), list()
    for idx, i in enumerate(playing_stat.HomeTeam):
        HTP.append(ht_points[i].pop(0))
    for idx, i in enumerate(playing_stat.AwayTeam):
        ATP.append(at_points[i].pop(0))
    playing_stat['HTP'] = HTP
    playing_stat['ATP'] = ATP
```

```python
for i in tqdm(range(0, 21)):
    get_points(table[i])
```
